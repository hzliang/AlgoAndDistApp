```python
import tensorflow as tf
```

```python
a = tf.constant(2)
b = tf.constant(3)
x = tf.add(a, b)
```

```python
with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs', sess.graph)
    print(sess.run(x))
    
writer.close()
```
