```python
import pandas as pd
import numpy as np
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn import cross_validation, metrics
from sklearn.model_selection import GridSearchCV

import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 4
```

```python
train = pd.read_csv('/root/train.csv',sep=',')
train.drop(['Unnamed: 0'],axis=1, inplace=True)
test = pd.read_csv('/root/test.csv',sep=',')
test.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
target_names='label'
predictors = [x for x in train.columns if x not in [target_names]]
predictors
```

```python
param_test2 = {
    'max_depth':[6,8,9],
    'min_child_weight':[1,4,6]
}
gsearch2 = GridSearchCV(estimator = XGBClassifier( learning_rate=0.1, n_estimators=30, max_depth=9,
                                        min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
                                        objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27), 
                       param_grid = param_test2, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch2.fit(train[predictors],train[target_names])
```
